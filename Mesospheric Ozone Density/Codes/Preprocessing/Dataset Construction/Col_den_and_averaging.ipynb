```python
import pandas as pd
import numpy as np
```

```python
f=pd.read_csv("C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\data_csv\\3 data set containing all points\\7 Combined files\\combined_dataset.csv")
df=pd.DataFrame(f)
df
```

```python
df_new=pd.DataFrame()
df_new
```

```python
col_den=df["col_den"][0]
i=0
while i<len(df["event"])-1:
    j=i
    count=0
    L=np.zeros((1,len(df.columns)),dtype=float)
    while df["col_den"][j]==col_den:
        if j%1e5==0:
            print(j)
        for k in range(L.shape[1]):
            col=df.columns[k]
            if col=="O3_127":
                continue
            L[0][k]+=float(df[col][j])
        j+=1
        count+=1
    L=L/count
    col_den=df["col_den"][j]
    dt=L.tolist()
    df_new=df_new.append(dt)
    i=j
```

```python
df_new.to_csv("C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\data_csv\\3 data set containing all points\\7.2 Combined files (Type 2)\\combined_dataset_new.csv")
```
