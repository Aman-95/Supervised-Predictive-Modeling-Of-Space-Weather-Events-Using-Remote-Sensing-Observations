```python
import pandas as pd
```

```python
dates1=[]
for i in range(1,10,1):
    dates1.append(f"0{i}")
dates1.append("10")

dates2=[]
for i in range(11,21,1):
    dates2.append(f"{i}")
    
dates3=[]
for i in range(21,31,1):
    dates3.append(f"{i}")
```

```python
L1=[]
L2=[]
L3=[]
for date in dates1:
    f=pd.read_csv(f"C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\data_csv\\collecting data in required altitude range\\5 with O3 density and log O3 density\\{date}_with_O3den.csv")
    df=pd.DataFrame(f)
    L1.append(df)
for date in dates2:
    f=pd.read_csv(f"C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\data_csv\\collecting data in required altitude range\\5 with O3 density and log O3 density\\{date}_with_O3den.csv")
    df=pd.DataFrame(f)
    L2.append(df)
for date in dates3:
    f=pd.read_csv(f"C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\data_csv\\collecting data in required altitude range\\5 with O3 density and log O3 density\\{date}_with_O3den.csv")
    df=pd.DataFrame(f)
    L3.append(df)
combined_file_1 = pd.concat(L1, axis=0, ignore_index=True)
combined_file_2 = pd.concat(L2, axis=0, ignore_index=True)
combined_file_3 = pd.concat(L3, axis=0, ignore_index=True)
combined_file_1.to_csv(f"C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\data_csv\\collecting data in required altitude range\\6 combined dataset\\combined_dataset_1.csv")
combined_file_2.to_csv(f"C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\data_csv\\collecting data in required altitude range\\6 combined dataset\\combined_dataset_2.csv")
combined_file_3.to_csv(f"C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\data_csv\\collecting data in required altitude range\\6 combined dataset\\combined_dataset_3.csv")
```
