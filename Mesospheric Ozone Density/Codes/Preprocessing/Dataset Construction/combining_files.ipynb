```python
import pandas as pd
import numpy as np
```

```python
dates=[]
time="Nov 2003"
for i in range(1,10):
    dates.append(f"0{i}")
for i in range(10,31):
    dates.append(f"{i}")
    
if time=="Nov 2003":
    dates.remove("18")
elif time == "Oct 2003":
    dates=[]
    for i in range(26,32):
        dates.append(f"{i}")
```

```python
L=[]
for date in dates:
    f=pd.read_csv(f"C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\data_csv\\{time}\\7 Adding Date\\{date}.csv")
    df=pd.DataFrame(f)
    df.drop("Unnamed: 0", inplace=True, axis=1)
    L.append(df)
combined_file = pd.concat(L, axis=0, ignore_index=True)
combined_file.to_csv(f"C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\data_csv\\{time}\\8 Combined dataset\\{time}.csv")
```
