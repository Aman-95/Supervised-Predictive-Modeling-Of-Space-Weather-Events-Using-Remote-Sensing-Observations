```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
f=pd.read_csv("C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\final_dataset_1.csv")
df=pd.DataFrame(f)
df
```

```python
L=[]
for col1 in df.columns:
    if col1 in ["O3_127","density","O3_96"]:
        continue
    l=[]
    for col2 in df.columns:
        if col2 in ["O3_127","density","O3_96"]:
            continue
        c=df[f"{col1}"].corr(df[f"{col2}"])
        l.append(c)
    L.append(l)
```

```python
Dic={"0":"event","1":"solAp","2":"solKp","3":"solf10p7Daily","4":"time",
     "5":"tpaltitude","6":"tplatitude","7":"tplongitude","8":"tpSolarZen","9":"ktemp",
    "10":"dst_index","11":"ae_index","12":"sym_h","13":"O3_density"}
```

```python
df2=pd.DataFrame(L,index=list(Dic.values()),columns=list(Dic.values()))
df2
```

```python
arr=np.array(L)
arr
```

```python
fig = plt.figure(figsize = (10,10))
sns.heatmap(arr, vmax = 1, square = False,xticklabels=list(Dic.values()), yticklabels=list(Dic.values()))
plt.show()
```
