```python
import pandas as pd
```

```python
L=[]
dates=["06","07","08","09","10","11","12"]

for date in dates:
    f1=pd.read_csv(f"C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\data_csv\\Final data\\final_with_dst_ae_symh\\final_{date}_with_dst_ae_symh.csv")
    df=pd.DataFrame(f1)
    D={}
    for col in df.columns:
        D[col]=0
    for i in range(len(df["event"])):
        for col in df.columns:
            if df[col][i]=="--" or df[col][i]==None:
                D[col]=D[col]+1
    L.append(D)
```

```python
for i in range(len(L)):
    print(f"For data file final_{dates[i]}:")
    for col in df.columns:
        print(col,"\t",L[i][col],"\t",'%.2f'%(100*(D[col]/len(df["event"]))),"%")
    print("\n\n")
```
