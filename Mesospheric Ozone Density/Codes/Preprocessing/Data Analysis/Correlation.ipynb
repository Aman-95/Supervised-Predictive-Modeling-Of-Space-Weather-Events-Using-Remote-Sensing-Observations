```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
f1=pd.read_csv("C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\data_csv\\collecting data in required altitude range\\6 combined dataset\\combined_dataset_1.csv")
f2=pd.read_csv("C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\data_csv\\collecting data in required altitude range\\6 combined dataset\\combined_dataset_2.csv")
f3=pd.read_csv("C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\data_csv\\collecting data in required altitude range\\6 combined dataset\\combined_dataset_3.csv")
df1=pd.DataFrame(f1)
df2=pd.DataFrame(f2)
df3=pd.DataFrame(f3)

df=pd.concat([df1,df2,df3])
df
```

```python
df.columns
```

```python
L=[]
for col1 in df.columns:
    if col1 in ["O3_127",'density', 'O3_96']:
        continue
    l=[]
    for col2 in df.columns:
        if col2 in ["O3_127",'density', 'O3_96']:
            continue
        c=df[f"{col1}"].corr(df[f"{col2}"])
        l.append(c)
    L.append(l)
```

```python
Dic={"0":"event","1":"solAp","2":"solKp","3":"solf10p7Daily","4":"time",
     "5":"tpaltitude","6":"tplatitude","7":"tplongitude","8":"tpSolarZen","9":"ktemp",
    "10":"ae_index","11":"dst_index","12":"sym_h","13":"O3_density","14":"log_O3_density"}
```

```python
df3=pd.DataFrame(L,index=list(Dic.values()),columns=list(Dic.values()))
df3
```

```python
arr=np.array(L)
```

```python
fig = plt.figure(figsize = (5,5))
sns.heatmap(arr, vmax = 1, square = False,xticklabels=list(Dic.values()), yticklabels=list(Dic.values()))
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(arr,vmin=-1.0, vmax=1.0,cmap="PiYG", center=0, fmt='.2f',square=True, linewidths=.5, annot=True, cbar_kws={"shrink": 0.80},xticklabels=df3.columns, yticklabels=df3.columns)
plt.show();
```
