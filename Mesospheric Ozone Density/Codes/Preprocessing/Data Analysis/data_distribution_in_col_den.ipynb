```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
f=pd.read_csv("C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\data_csv\\3 data set containing all points\\7 Combined files\\combined_dataset.csv")
df=pd.DataFrame(f)
```

```python
L=np.zeros((2,11),dtype=int)
colden=df["col_den"]
length=len(colden)
for i in range(length):
    x=colden[i]
    if i%10000==0:
        print(f"{i}/{length}")
    x=x/1e9
    if x>10:
        x=x/10
        if x>=10:
            L[1][10]+=1
        else:
            L[1][int(x)]+=1
    elif x==10:
        L[0][10]+=1
        L[1][1]+=1
    else:
        L[1][0]+=1
        L[0][int(x)]+=1
```

```python
L
```

```python
print("for Column density")
print("\n\nIn range 0*10^9-10*10^9")
for i in range(11):
    print(f"{i}    {round(100*L[0][i]/sum(L[0]),4)}%    {round(100*L[0][i]/(sum(L[0])+sum(L[1])),4)}%")
print("\n\nIn range 0*10^10-10*10^10")
for i in range(0,11):
    print(f"{i}    {round(100*L[1][i]/sum(L[1]),4)}%    {round(100*L[1][i]/(sum(L[0])+sum(L[1])),4)}%")
```

```python
C=[]
for i in range(len(L[0])):
    C.append(L[0][i])
for i in range(1,len(L[1])):
    C.append(L[1][i])
C
C_=[]
for i in range(0,10):
    C_.append(f"{i}")
for i in range(10,101,10):
    C_.append(f"{i}")
C_.append("~")
C_
```

```python
plt.plot(100*L[0]/sum(L[0]))
plt.plot(100*L[0]/(sum(L[0])+sum(L[1])))
plt.scatter(range(11),100*L[0]/sum(L[0]))
plt.title("column density distribution")
plt.ylabel("No. of data points in %")
plt.xlabel("column density range (x*10^9)")
plt.legend(['Overall','local'],loc='upper right')
```

```python
plt.plot(100*L[1]/sum(L[1]))
plt.scatter(range(11),100*L[1]/sum(L[1]))
plt.plot(100*L[1]/(sum(L[0])+sum(L[1])))
plt.title("column density distribution")
plt.ylabel("No. of data points in %")
plt.xlabel("column density range (x*10^10)")
plt.legend(['Overall','local'],loc='upper right')
```
