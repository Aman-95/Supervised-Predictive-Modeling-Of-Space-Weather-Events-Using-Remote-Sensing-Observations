```python
f1=pd.read_csv('C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\WDC data\\Oct 2003\\A{X} index.txt',sep=",",header=None)
f1
```

```python
D=[]
df=pd.DataFrame()
i=0
for l in f1[0]:
    l=l.replace("    ",",")
    l=l.replace("   ",",")
    l=l.replace("  ",",")
    l=l.replace(" ",",")
    l=l.replace(",,,",",")
    l=l.replace(",,",",")
    l=l.replace(", ",",")
    L=l.split(",")
    D.append(L)
    if(len(L)!=64):
        print(f"{i}  {len(L)}")
    i=i+1
df=df.append(D)
df.to_csv("C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\WDC data\\Oct 2003\\A{X} index.csv")
```
