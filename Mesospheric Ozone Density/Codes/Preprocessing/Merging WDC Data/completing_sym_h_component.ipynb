```python
import pandas as pd
import numpy as np
```

```python
f1=pd.read_csv("C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\Other data\\empty.csv")
df1=pd.DataFrame(f1)
df1
```

```python
L=[]
for i in range(len(df1["head"])):
    s=df1["head"][i]
    s=s.strip("[")
    s=s.strip("]")
    s=s.replace("     ","&")
    s=s.replace("    ","&")
    s=s.replace("   ","&")
    s=s.replace("  ","&")
    s=s.replace(" ","&")
    s=s.replace("\n","")
    s=list(s.split("&"))
    L.append(s)
```

```python
df=pd.DataFrame(L)
df
```

```python
df.to_csv("C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\Other data\\rest_sym_h.csv")
```
