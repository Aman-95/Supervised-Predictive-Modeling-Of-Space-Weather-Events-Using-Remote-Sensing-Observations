```python
f1=pd.read_csv("C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\Other data\\empty.csv")
df1=pd.DataFrame(f1)
df1
```

```python
L=[]
for i in range(49,73,1):
    s=df1["head"][i]
    s=s.strip("[")
    s=s.strip("]")
    s=s.replace("     ","&")
    s=s.replace("    ","&")
    s=s.replace("   ","&")
    s=s.replace("  ","&")
    s=s.replace(" ","&")
    s=s.replace("\n","")
    s=list(s.split("&"))
    L.append(s)
```

```python
df=pd.DataFrame(L)
df
```
