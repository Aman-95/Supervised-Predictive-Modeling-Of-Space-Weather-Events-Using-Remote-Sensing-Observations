```python
D=[]
df=pd.DataFrame()
i=0
for l in f1[0]:
    l=l.replace("    ",",")
    l=l.replace("   ",",")
    l=l.replace("  ",",")
    l=l.replace(" ",",")
    l=l.replace(",,,",",")
    l=l.replace(",,",",")
    l=l.replace(", ",",")
    L=l.split(",")
    D.append(L)
    if(len(L)!=63):
        print(f"{i}  {len(L)}")
    i=i+1
df=df.append(D)
df.to_csv("C:\\Users\\Acer\\PycharmProjects\\BTech_Project\\WDC data\\Nov 2003\\ASYM_SYM index.csv")
```
